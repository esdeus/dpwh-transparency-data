Let's begin cleaning the data

```python
df.rename(columns={
    "location.province": "province",
    "location.region": "region"
}, inplace=True)
```

```python
# Convert timestamps to real dates
df['startDate'] = pd.to_datetime(df['startDate'], unit='ms', errors='coerce')
df['completionDate'] = pd.to_datetime(df['completionDate'], unit='ms', errors='coerce')
```

```python
# Drop empty columns (livestreamUrl, livesteamVideoId, livesteamDetectedAt)
df.drop(columns=[
    "livestreamUrl",
    "livestreamVideoId",
    "livestreamDetectedAt",
], inplace=True)
```

Now for columns with missing data

```python
text_columns = [
    "componentCategories",
    "contractor",
    "programName",
    "sourceOfFunds",
]

df[text_columns] = df[text_columns].fillna("No data available")
```

```python
# Standardize the column names
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
# Remove duplicates
df = df.drop_duplicates(subset=['contractid'])
```

```python
# For quality checking of the team
df.to_csv("dpwh_cleaned_data.csv", index=False)
```

```python
# For connecting the data to the dashboard
df.to_json("dpwh_cleaned_data.json", orient="records")
```
